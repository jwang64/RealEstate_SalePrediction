```python
import pandas as pd
```

```python
initialData = pd.read_csv('train.csv')
```

```python
initialData = initialData.set_index('Id')
```

### Dropping Columns with mostly missing data

```python
def drop_missing_threshhold(dataFrame,threshold):
    dataFrameSize = dataFrame.shape[0]
    missing_list = [c for c in dataFrame.columns if sum(dataFrame[c].isnull()) >= threshold*dataFrameSize]
    dataFrame = dataFrame.drop(columns=missing_list,axis=1)
    return dataFrame
```

```python
updatedDataset = drop_missing_threshhold(initialData,.3)
```

```python
updatedDataset
```

### Removing nonpredictors of sale price

```python
updatedDataset = updatedDataset.drop(columns=['SaleCondition','SaleType','YrSold','MoSold'])
```

### Handling Missing Data

```python
updatedDataset['LotFrontage'].mean()
```

```python
updatedDataset['LotFrontage'] = updatedDataset['LotFrontage'].fillna(updatedDataset['LotFrontage'].mean())
```

```python
missing_list_df = updatedDataset.isna().sum(axis=0).sort_values(ascending=False).rename_axis('columnList').reset_index()
```

```python
missing_list = missing_list_df[missing_list_df[0] > 1]['columnList'].to_list()
```

```python
updatedDataset
```

```python
updatedDataset = updatedDataset.dropna(subset=missing_list,how='any')
```

### Ordinal Data

```python
def ordinalRatings(value):
    if (value == 'Ex'):
        return 5
    elif (value == 'Gd'):
        return 4
    elif (value == 'Av'):
        return 3
    elif (value == 'TA'):
        return 3
    elif (value == 'Fa'):
        return 2
    elif (value == 'Po'):
        return 1
    else:
        return 0
```

```python
def ordinalRatingBasement(value):
    if (value == 'GLQ'):
        return 5
    elif (value == 'ALG'):
        return 4
    elif (value == 'Rec'):
        return 4
    elif (value == 'BLG'):
        return 3
    elif (value == 'LwQ'):
        return 2
    elif (value == 'Unf'):
        return 1
    else:
        return 0
```

```python
updatedDataset['HeatingQC'] = updatedDataset.apply(lambda x: ordinalRatings(x['HeatingQC']),axis=1)
updatedDataset['KitchenQual'] = updatedDataset.apply(lambda x: ordinalRatings(x['KitchenQual']),axis=1)
updatedDataset['GarageQual'] = updatedDataset.apply(lambda x: ordinalRatings(x['GarageQual']),axis=1)
updatedDataset['GarageCond'] = updatedDataset.apply(lambda x: ordinalRatings(x['GarageCond']),axis=1)
updatedDataset['BsmtQual'] = updatedDataset.apply(lambda x: ordinalRatings(x['BsmtQual']),axis=1)
updatedDataset['ExterQual'] = updatedDataset.apply(lambda x: ordinalRatings(x['ExterQual']),axis=1)
updatedDataset['ExterCond'] = updatedDataset.apply(lambda x: ordinalRatings(x['ExterCond']),axis=1)
updatedDataset['BsmtCond'] = updatedDataset.apply(lambda x: ordinalRatings(x['BsmtCond']),axis=1)
updatedDataset['BsmtExposure'] = updatedDataset.apply(lambda x: ordinalRatings(x['BsmtExposure']),axis=1)
```

```python
updatedDataset['BsmtFinType1'] = updatedDataset.apply(lambda x: ordinalRatingBasement(x['BsmtFinType1']),axis=1)
updatedDataset['BsmtFinType2'] = updatedDataset.apply(lambda x: ordinalRatingBasement(x['BsmtFinType2']),axis=1)
```

### Feature Engineering

```python
updatedDataset['totalSF'] = updatedDataset.apply(lambda x: x['TotalBsmtSF'] + x['1stFlrSF'] + x['2ndFlrSF'] + x['WoodDeckSF'] + x['OpenPorchSF'], axis =1)
```

```python
updatedDataset['interiorSF'] = updatedDataset.apply(lambda x: x['TotalBsmtSF'] + x['1stFlrSF'] + x['2ndFlrSF'], axis =1)
```

```python
updatedDataset['BathroomNum'] = updatedDataset.apply(lambda x: x['BsmtFullBath'] + (.5 *x['BsmtHalfBath']) + x['FullBath'] + (.5 * x['HalfBath']), axis=1)
```

### Encoding Categorical Variables

```python
updatedDataset = pd.get_dummies(updatedDataset)
```

### Saving as CSV

```python
updatedDataset.to_csv('cleaned_realestate.csv')
```
